```python
import pandas as pd 
from sklearn import tree
import numpy as np
from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV
```

```python
d = pandas.read_csv("C:/Users/affsa/training/day 1/DIATM 2019/wine.data",names=["Class","Alcohol","Malic Alic","Ash","Alcanility of Ash","Magnesium",
                    "Total phenols","Flavanoids","Nonflavanoid phenols","Proanthocyanins","Color intensity","Hue","OD280_OD315_diluted wines",
                    "Proline"])

y = d["Class"].values
del(d["Class"])
```

```python
clh = tree.DecisionTreeClassifier(max_depth  = 10)
clf = AdaBoostClassifier(base_estimator= clh,n_estimators=10)
clf.fit(d.values, y)
```

```python
print(np.mean(cross_val_score(clf, d.values, y, cv=5)))
```

```python
clh = tree.DecisionTreeClassifier(max_depth  = 10)
parameters = {'max_depth':np.arange(1,20)}
```

```python
clfx = GridSearchCV(clh, parameters)
```

```python
clfx.fit(d.values, y)
```

```python
print(clfx.best_params_)
```

```python
clh = tree.DecisionTreeClassifier(max_depth  = 3)
clf = AdaBoostClassifier(base_estimator= clh,n_estimators=10)
clf.fit(d.values, y)
```

```python
print(np.mean(cross_val_score(clf, d.values, y, cv=5)))
```
